# Imports


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Input Section

```python
# here we will take the input of the actors name
actor_name = input('\t Enter the name of the actor')

# changing the name into desired form
actor_name= actor_name.replace(' ', '_')

actor_name
```

```python
url = f'https://en.wikipedia.org/wiki/{actor_name}_filmography'
```

# fetching the data


```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup.prettify()
```

```python
table = soup.find('table', class_='wikitable sortable plainrowheaders')
table.find_all('tr')
```

```python
# movies list 
movies_list =[]

for rows in table.find_all('tr')[2:]:
    row =rows.find('td').text.strip('\n')
    movies_list.append(row)
```

```python
# now descending order
movies_list = movies_list[::-1]
# final result
print(f'The Descending order of movies list of {actor_name} is:')
for movie in movies_list:
    print(movie)
```
